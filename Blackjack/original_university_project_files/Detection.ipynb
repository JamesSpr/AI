```python
# https://www.youtube.com/watch?v=bYBFF-N1Z98&ab_channel=PyOhio - Card Detection
# https://www.youtube.com/watch?v=Tm_7fGolVGE&ab_channel=Murtaza%27sWorkshop-RoboticsandAI - Warp Perspective
# https://github.com/geaxgx/playing-card-detection/blob/master/creating_playing_cards_dataset.ipynb - Example case

import cv2
import numpy as np

# Frame width and height are dimensions after cropping
frameW = 1180
frameH = 1000

# Card orientation arrays for transformation to width and height
width, height = 200, 250
normalOrientation = np.float32([[0,0], [width, 0], [width, height], [0, height]])
rotatedOrientation = np.float32([[width, 0], [width, height], [0, height], [0,0]])

# Values for corner reference
cxMin = 2
cxMax = 35
cyMin = 3
cyMax = 65


class Card:
    def __init__(self, x, y, w, h, contour, warp):
        self.x = x
        self.y = y
        self.w = w
        self.h = h
        self.contour = contour
        self.warp = warp
        self.value = ""
        self.suit = ""

# Preprocess Image
def get_edges(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.bilateralFilter(gray, 10, 15, 15)
    edges = cv2.Canny(blur, 50, 150, False)
    return edges

def dilate_img(img):
    kernel = np.ones((2,2), np.uint8)
    dilate = cv2.dilate(img.copy(), kernel, iterations=1)
    return dilate

# Gets the external contours of an image
def get_contours(img):
    contours, _ = cv2.findContours(img.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    sorted(contours, key=cv2.contourArea, reverse=True)
    return contours

def transform_cards(contours, frame):
    cards = []
    for c in contours:
        area = cv2.contourArea(c)
        if area > 36000 and area < 42000:
            rect = cv2.minAreaRect(c)
            box = cv2.boxPoints(rect)
            box = np.int0(box)
            # Get dimensions and coordinates of box
            ((x,y), (w,h), a) = rect

            if w > h:
                M = cv2.getPerspectiveTransform(np.float32(box), normalOrientation)
            else:
                M = cv2.getPerspectiveTransform(np.float32(box), rotatedOrientation)
                
            warp = cv2.warpPerspective(frame, M, (width, height))
            cards.append(Card(x, y, w, h, c, warp))

    return cards

def show_preprocessed_frames():
    # Get Video Stream
    cap = cv2.VideoCapture(0)

    while(True):
        # Capture each frame and crop to usable area
        ret, frame = cap.read()
        frame = frame[50:1050, 400:1580, :] # for 1080p

        # Perform Feature extraction to get individual cards
        edges = get_edges(frame)
        dilate = dilate_img(edges)
        contours = get_contours(dilate)
        contoursImg = cv2.drawContours(frame.copy(), contours, -1, (0, 255,0 ), 2)
        cards = transform_cards(contours, frame)
    
        # Display Video Feed with contours
        cv2.imshow("Video Feed", frame)
        cv2.imshow("Edges", edges)
        
        if cv2.waitKey(1) == 27:
            break

    cap.release()
    cv2.destroyAllWindows()
    
def show_all_cards_in_frame():
    # Hands
    dealer = []
    player = []

    # Get Video Stream
    cap = cv2.VideoCapture(0)

    while(True):
        # Capture each frame and crop to usable area
        ret, frame = cap.read()
        frame = frame[50:1050, 400:1580, :] # for 1080p

        # Perform Feature extraction to get individual cards
        edges = get_edges(frame)
        dilate = dilate_img(edges)
        contours = get_contours(dilate)
        contoursImg = cv2.drawContours(frame.copy(), contours, -1, (0, 255,0 ), 2)
        cards = transform_cards(contours, frame)

        # Display Video Feed with contours
        cv2.imshow("Video Feed", contoursImg)

        # Show each card in video feed
        i = 0
        for card in cards:
            cv2.imshow("Card" + str(i), card.warp)
            i += 1
            # Assign card to player depending on location, top of screen is dealer
            if card.y < frameW/2:
                dealer.append(card)
            else:
                player.append(card)

        if cv2.waitKey(1) == 27:
            break

    cap.release()
    cv2.destroyAllWindows()
    
    
def start():   
    # Get Video Stream
    cap = cv2.VideoCapture(0)

    while(True):
        # Hands
        dealer = []
        player = []
        
        # Capture each frame and crop to usable area
        ret, frame = cap.read()
        frame = frame[50:1050, 400:1580, :] # for 1080p
        output = frame.copy()
        # Perform Feature extraction to get individual cards
        edges = get_edges(frame)
        dilate = dilate_img(edges)
        contours = get_contours(dilate)
        cards = transform_cards(contours, frame)

        # Assign each card to a player, depending on the position in frame
        for card in cards:
            if card.y < frameW/2:
                dealer.append(card)
            else:
                player.append(card)
        
        for card in dealer:
            value, suit = classify_card(card.warp)
            output = cv2.putText(output, (str(value) + str(suit)),(int(card.x)-15,int(card.y)), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 3)

        for card in player:
            value, suit = classify_card(card.warp)
            output = cv2.putText(output, (str(value) + str(suit)),(int(card.x)-15,int(card.y)), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 3)
        
        # Display Video Feed with contours
        cv2.imshow("Video Feed", output)
        
        if cv2.waitKey(1) == 27:
            break

    cap.release()
    cv2.destroyAllWindows()
    return dealer, player

def play():   
    # Get Video Stream
    cap = cv2.VideoCapture(0)

    while(True):
        # Hands
        dealer = []
        player = []
        
        # Capture each frame and crop to usable area
        ret, frame = cap.read()
        frame = frame[50:1050, 400:1580, :] # for 1080p
        output = frame.copy()
        output = cv2.line(output, pt1=(0,500), pt2=(1180, 500), color=(0, 255, 0), thickness=5, lineType=8, shift=0)
        # Perform Feature extraction to get individual cards
        edges = get_edges(frame)
        dilate = dilate_img(edges)
        contours = get_contours(dilate)
        cards = transform_cards(contours, frame)

        # Assign each card to a player, depending on the position in frame
        for card in cards:
            if card.y < frameW/2:
                dealer.append(card)
            else:
                player.append(card)
        
        dealerHand = 0
        for card in dealer:
            value, suit = classify_card(card.warp)
            if value == "A":
                if dealerHand <= 10:
                    dealerHand += 11
                else:
                    dealerHand += 1
            elif value == "J":
                dealerHand += 10
            elif value == "Q":
                dealerHand += 10
            elif value == "K":
                dealerHand += 10
            else:
                dealerHand += int(value)
        output = cv2.putText(output, "Dealer Hand Value: " + str(dealerHand), (10,50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
        
        
        usableAce = False
        playerHand = 0
        for card in player:
            value, suit = classify_card(card.warp)
            if value == "A":
                usableAce = True
                if playerHand <= 10:
                    playerHand += 11
                else:
                    playerHand += 1
            elif value == "J":
                playerHand += 10
            elif value == "Q":
                playerHand += 10
            elif value == "K":
                playerHand += 10
            else:
                playerHand += int(value)
        output = cv2.putText(output, "AI Hand Value: " + str(playerHand), (10,950), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
        
        for key, value in QLearning_Actions.items():
            if key == (playerHand, dealerHand, usableAce):
                output = cv2.putText(output, "AI Turn: " + translate_value(value), (10,900), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
        
        if dealerHand >= 17:
            output = cv2.putText(output, "Game Over: " + get_outcome(playerHand, dealerHand), (150,475), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,0), 3)
        
        # Display Video Feed with contours
        cv2.imshow("Video Feed", output)
        
        if cv2.waitKey(1) == 27:
            break

    cap.release()
    cv2.destroyAllWindows()
```

```python
# Helper function to translate value into action
def translate_value(val):
    if val == 1:
        return "Hit"
    else:
        return "Stand"
    
# Helper function to determine outcome of game
def get_outcome(playerHand, dealerHand):
    if playerHand <= 21 and (dealerHand < playerHand or dealerHand > 21):
        return "AI Wins"
    elif dealerHand <= 21 and (playerHand < dealerHand or playerHand > 21):
        return "Dealer Wins"
    elif dealerHand > 21 and playerHand > 21:
        return "No Winner"
    elif dealerHand == playerHand:
        return "No Winner"
    else: 
        return "CHECK OUTCOME" # In case of edge cases - Testing
```
